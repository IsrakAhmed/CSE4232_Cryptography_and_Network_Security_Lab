6. Use the Lehmann algorithm to check whether the given number P is prime or not?

```python
import random

def lehmann_primality_test(N, k=5):
    if N <= 2:
        return N == 2

    if N % 2 == 0:
        return False

    for _ in range(k):
        a = random.randint(2, N - 2)
        exp = (N - 1) // 2
        x = pow(a, exp, N)

        if x != 1 and x != N - 1:
            return False

    return True  



N = int(input("Enter a number to test: "))


if lehmann_primality_test(N, k=10):
    print(f"{N} is probably prime.")
else:
    print(f"{N} is composite.")
```
