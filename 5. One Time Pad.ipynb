5. You are supplied a file of large nonrepeating set of truly random key letter. Your job is to encrypt the plaintext using ONE TIME PAD technique. Then perform the reverse operation to get original plaintext.

```python
def letter_to_num(c):
    return ord(c.upper()) - ord('A') + 1


def num_to_letter(n):
    n = n - 1 + 26
    return chr((n % 26) + ord('A'))


def encrypt(plain_text):
    cipher_text = ""
    idx = 0

    for ch in plain_text:
        p_val = letter_to_num(ch)
        k_val = letter_to_num(key[idx])
        x = (p_val + k_val) % 26
        
        if ch == ' ':
            cipher_text += ' '
        else:
            cipher_text += num_to_letter(x)
            idx += 1

    return cipher_text


def decrypt(cipher_text):
    plain_text = ""
    idx = 0
    
    for ch in cipher_text:
        c_val = letter_to_num(ch)
        k_val = letter_to_num(key[idx])
        x = (c_val - k_val + 26) % 26

        if ch == ' ':
            plain_text += ' '
        else:
            plain_text += num_to_letter(x)
            idx += 1

    return plain_text, idx



key = "abcdefghijklmnopqrstuvwxyz".upper()
plain_text = "computer science".upper()

cipher_text = encrypt(plain_text)
decrypted_text, idx = decrypt(cipher_text)

key = key[idx:]

print(f"Plain Text: {plain_text}")
print(f"Encrypted Text: {cipher_text}")
print(f"Decrypted Text: {decrypted_text}")
print(f"Remaining Key: {key}")
```
