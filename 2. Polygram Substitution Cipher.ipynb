2. Find out the Polygram Substitution Cipher of a given plaintext (Consider the block size of 3). Then perform the reverse operation to get original plaintext.

```python
encryption_rules, decryption_rules = {}, {}

substitution_map = {
    "HEL": "XQZ",
    "WOR": "MNP",
    "LOW": "YWA",
    "ORL": "STV"
}


encryption_rules, decryption_rules = {}, {}


for key, value in substitution_map.items():
    encryption_rules[key] = value
    decryption_rules[value] = key


def pad_text(text, blockSize=3):
    while len(text) % blockSize != 0:
        text += "?"
    return text


def encrypt(plaintext, blockSize=3):
    plaintext = pad_text(plaintext.upper(), blockSize)
    cipher = ""
    for i in range(0, len(plaintext), blockSize):
        block = plaintext[i:i+blockSize]
        cipher += encryption_rules.get(block, block)
    return cipher.rstrip("?")


def decrypt(ciphertext, blockSize=3):
    plain = ""
    for i in range(0, len(ciphertext), blockSize):
        block = ciphertext[i:i+blockSize]
        plain += decryption_rules.get(block, block)
    return plain.rstrip("?")


plaintext = "HELLO WORLD"
ciphertext = encrypt(plaintext)
decrypted = decrypt(ciphertext)

print("Plaintext :", plaintext)
print("Ciphertext:", ciphertext)
print("Decrypted :", decrypted)
```
